```python
from pitchplots.static import tonnetz
import matplotlib.pyplot as plt
import glob
```

```python
files = glob.glob("data/*.csv")
files
```

```python
kwargs = {
    "colorbar":False,
    "figsize":(10,10),
    "fontsize":2,
    "cmap":'Reds',
    "nan_color":'w',
    "duplicate":True
}
```

```python
import pandas as pd
csv = pd.read_csv(files[11], index_col=0)
csv['tpc'] = csv['tpc'].str.replace('x', '##')
csv.tpc.unique()
```

```python
fig = tonnetz(
    csv, 
    radius=5,
    **kwargs
    )

fig.savefig('../ExtendedTonality/thesis/img/macro/Liszt_Lugubre_gondola_I_200_1_tonnetz.png', dpi=300)
```
